En un histograma, visualizar el GDP per Capita de cada país. Bonus: ordenar el histograma en orden decreciente.



```python
import pandas as pd
import matplotlib.pyplot as plt
w= pd.read_csv('dictionary.csv', low_memory=False)
tabla1=w['Country'].copy()
tabla2=w['GDP per Capita'].copy()
nueva_tabla= pd.concat([tabla1, tabla2], axis=1)
plt.figure();
nueva_tabla.plot.hist()
```

¿Cuál es el número de medalla por cada país, sin diferenciar el tipo de medalla, en los juegos de verano? Visualizar en un histograma.


```python
import pandas as pd
import matplotlib.pyplot as plt
s= pd.read_csv('summer.csv', low_memory=False)
counts = s.groupby(['Country'],axis=0).sum()
plt.figure();
counts.plot.hist()
```

¿Cuál es el número de medalla por cada país, diferenciando el color de la medalla, en los juegos de verano? Visualizar en un histograma.

```python
import pandas as pd
import matplotlib.pyplot as plt
s= pd.read_csv('summer.csv', low_memory=False)
counts = s.groupby(['Country','Medal']).aggregate(sum)
plt.figure();
counts.plot.hist()
```
